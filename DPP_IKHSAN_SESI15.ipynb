```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a dataframe from the given data
data = {'Jurusan': ['TI', 'SI', 'AK', 'TE', 'MJ', 'HK', 'DKV', 'PGSD', 'MS'],
        'Jumlah': [1000, 800, 1100, 500, 2000, 400, 300, 500, 200]}
df = pd.DataFrame(data)

# Plot a bar chart of the number of students for each department
df.plot(x='Jurusan', y='Jumlah', kind='bar')
plt.title('Number of Students by Department')
plt.xlabel('Department')
plt.ylabel('Number of Students')
plt.show()
```
